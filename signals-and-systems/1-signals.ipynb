# Continuous Signals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
t0 = 0
tn = 2 * np.pi
n = 50
w = 1
```

```python
# time vector
dt = (tn-t0)/n
t = np.arange(t0, tn, dt)
```

```python
# signal
signal = np.sin(w * t)
```

```python
# output
print('time: ')
print(t)
print('signal: x(t) ')
print(signal)
```

```python
# graphs
plt.plot(t, signal)
plt.xlabel('t')
plt.ylabel('signal x[(t)]')
plt.show()
```
